````python
#Начало вашего кода
import csv

def load_csv(filename=None):
    if os.path.exists(filename):
        data_lst = []
        with open(filename) as f:
            csv_data = csv.reader(f)
            
            
            

            for row in csv_data:
                
                data_lst.append(row)
                
            header = data_lst[0]
            data = data_lst[1:]
             
                
            return header,data
        
       
    else:
        print('./SocialNetworkAds.csv does not exist')
        header = None
        data  = None
        
        return header,data

#Конец вашего кода
header,data=load_csv("Social_Network_Ads.csv")
print(header)
for i in range(5):
    print(data[i])
````

### Шаг 3

Пожалуйста, реализуйте функцию  ```statistics(data, idx)```, которая вычисляет базовую статистику для данного столбца ```idx``` в ```ndarray``` ```data```. Функция должна возвращать словарь рассчитанных значений.

````python
#Начало вашего кода
def statistics(data=None, idx=None):
    
    
    list_of_genders = [row[idx] for row in data]
    
    myDict = ('mean','std','median','max','min')
    
    out_num = np.mean(list_of_genders),np.std(list_of_genders),np.median(list_of_genders),np.max(list_of_genders),np.min(list_of_genders)

    
    return dict(zip(myDict, out_num))

#Конец вашего кода
age_stats=statistics(np_data,2)
salary_stats=statistics(np_data,3)

print(age_stats)
print(salary_stats)
````
