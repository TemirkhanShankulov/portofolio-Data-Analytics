````python
def random_crop(img, width=None, height=None, seed=1):
    np.random.seed(seed) 
    x = np.random.randint(img.shape[1]-height)
    
    y = np.random.randint(img.shape[0]-width)
    
    crop = img[y:height+y, x:width+x]

    plt.axis("off")

    return crop

plt.imshow(random_crop(img_cv,width=100,height=100,seed=3));
#Начало вашего кода

#Конец вашего кода
````

Важно отметить, что ```seed``` изменяется при каждом вызове функции ```random_crop()```, начиная с ```1``` и заканчивая ```num_rows * num_cols```.
Пожалуйста, посмотрите на ожидаемый результат для лучшего понимания.

````python
def plot_random_grid(img,num_rows=2,num_cols=2,width=None, height=None):
    

    fig, axs = plt.subplots(num_rows, num_cols)

    seed=1

    for i in range(num_rows):
        for j in range(num_cols):

            axs[i][j].imshow( random_crop(img,width=width,height=height,seed=seed))
            
            axs[i][j].set_title("Seed: %s (100x100)" % (seed))
            
            axs[i,j].tick_params(axis  = 'both', bottom= False, left  = False,
                             labelbottom = False, labelleft   = False)
            
#             axs[i][j].set_xticks([],[])
#             axs[i][j].set_yticks([],[])
            
            seed += 1
            plt.axis("off")
            
plot_random_grid(img_cv,num_rows=2,num_cols=3,width=100,height=100)    
````

````python
def plot_grid(img,num_rows=2,num_cols=2):
    #Начало вашего кода
    fig, axs = plt.subplots(num_rows, num_cols)


#     rows_lst = []
#     columns_lst = []
    height,width,_=img.shape
    h = height//num_rows

    w = width//num_cols
  
    
    starty=0
    endy=h

    for i in range(0,num_rows):
        startx=0
        endx=w

        for j in range(0,num_cols):
            new_img=img[starty:endy,startx:endx,:]
            axs[i][j].imshow(new_img)
            axs[i][j].set_xticks([])
            axs[i][j].set_yticks([])
            startx+=w
            endx+=w

        starty+=h
        endy+=h

  

    #Конец вашего код`а

plot_grid(img_cv,3,2)
````
