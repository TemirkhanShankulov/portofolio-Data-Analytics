Пожалуйста, реализуйте функцию ```plot_sns_age_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
# titanic[titanic['pclass']==1]
````

````python
def plot_age_pclass(data=None):
    #Начало вашего кода
    plt.figure(figsize=(4,8));
    classes = set(data['pclass']);
    for c in classes:
        sns.catplot(data=titanic[titanic['pclass']==c], x='sex', y='age', hue='alive', hue_order=['yes','no'],row='pclass', height=4, aspect=2).set(title=f'Passenger class: {c}');    
    #Конец вашего кода
plot_age_pclass(data=titanic)
````

### Шаг 7

Пожалуйста, реализуйте функцию plot_counts(data=None), которая отображает несколько фигур, каждая фигура содержит визуализацию на основе ```param```.

````python
def plot_counts(data=None,param=[],fontsize=15):
    #Начало вашего кода
    plt.figure(figsize=(4,8))
#     fig, axs = plt.subplots(nrows=1,ncols=2)

    sns.set_style('whitegrid')
    sns.set_context('notebook')
  
    f, axes = plt.subplots(2,1)
   
    for i,j in enumerate(param):
        
        sns.countplot(x='alive', hue=j, data=data, ax=axes[i]).set_title('Counts based on {}'.format(j), fontsize=fontsize);
    
    
    f.tight_layout()

plot_counts(data=titanic,param=["pclass","sex"])
````

````python
def plot_dists(data=None,param=[],fontsize=15):
    #Начало вашего кода
    plt.figure(figsize=(4,4))
    sns.set_context("notebook")
    sns.set_style('whitegrid')
  
    f, axes = plt.subplots(2,1)
    for i,j in enumerate(param):
        sns.distplot(titanic[j], ax=axes[i]).set_title('{} distribution'.format(j), fontsize=fontsize)
        
        
    f.tight_layout()
        


plot_dists(data=titanic,param=["fare","age"])
````
