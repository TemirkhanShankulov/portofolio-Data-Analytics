### Шаг 3

### Ожидаемый результат:



````python
drop_data = titanic.drop(columns=['adult_male','alone'])
````

````python
#Начало вашего кода
sns.set(font_scale=1.5)
sns.pairplot(drop_data,diag_kind='kde',hue='sex',corner=True);
#Конец вашего кода
````

### Шаг 4

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода

sns.set(font_scale=1.5)
h=sns.FacetGrid(titanic,row='alone',col='sex',height=5,hue='alive')
h.map_dataframe(sns.scatterplot,'age','fare').add_legend();
#Конец вашего кода
````

### Шаг 5

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода
sns.set(font_scale=1.5)
sns.set_style('darkgrid')
h=sns.FacetGrid(titanic,row='alone',col='pclass',height=5,hue='alive')
h.map_dataframe(sns.swarmplot,'sex','age').add_legend();
#Конец вашего кода
````
