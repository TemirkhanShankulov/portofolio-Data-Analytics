3. Пожалуйста, обработайте данные так, чтобы у вас был следующий DataFrame. Преобразуйте категориальные данные в целые числа.



````python
pd_data=exercise.copy()
# Начало вашего кода

diet_int = []
time_int = []
kind_int = []
for i in pd_data['diet']:
    if i=='low fat':
        diet_int.append(0)
    else:
        diet_int.append(1)
        
for i in pd_data['time']:
    if i=='1 min':
        time_int.append(0)
    elif i=='15 min':
        time_int.append(1)
    else:
        time_int.append(2) 
        
for i in pd_data['kind']:
    if i=='rest':
        kind_int.append(0)
    elif i=='walking':
        kind_int.append(1)
    else:
        kind_int.append(2)       
        
        
pd_data['diet_int'] = diet_int
pd_data['time_int'] = time_int 
pd_data['kind_int'] = kind_int 

pd_data = pd_data.drop("Unnamed: 0",axis=1)
# Конец вашего кода
pd_data.head()
````

9. Пожалуйста, произведите следующий график, показанный ниже:



````python
flights_data=flights.copy()
````

````python
from datetime import datetime
lst=[]
for i in flights_data['month']:
    mname = i
    mnum = datetime.strptime(mname, '%b').month
    result = str(mnum)
    lst.append(result)

````

````python
import datetime
lst_month = []
for i in lst:
    datetime_object = datetime.datetime.strptime(i, "%m")

    
    full_month_name = datetime_object.strftime("%B")
    lst_month.append(full_month_name)
````

````python
flights_data['month'] = lst_month
````

````python
flights_newdata = flights_data
````

````python
# Начало вашего кода
sns.set_style('darkgrid')
h=sns.FacetGrid(flights_newdata,col='month',col_wrap=4)
h.map_dataframe(sns.scatterplot,'year','passengers');
plt.xticks([1950,1955,1960],['1950','1955','1960']);
plt.yticks([200,400,600],['200','400','600']);
# Конец вашего кода
````
