### Шаг 2

Реализуйте функцию ```plot_corr_matrix(data,figsize=(30,15))```, которая будет вычислять коэффициенты корреляции для входного аргумента ```data``` и построит тепловую карту, используя ```matplotlib.pyplot```, как показано в ожидаемом результате.

**Примечание:** Размер шрифта заголовка установлен на 15.

````python
def plot_corr_matrix(data,figsize=(30,15)):
    #Начало вашего кода
     
#     print(data.columns.values)
    corr_matrix = data.corr()
#     return corr_matrix
    plt.rcParams["figure.figsize"] = figsize
    fig, ax = plt.subplots();
    im = ax.imshow(corr_matrix)
    
    plt.xticks(np.arange(len(data.columns.values)), labels=data.columns.values)
    plt.xticks(rotation=90)
    plt.yticks(np.arange(len(data.columns.values)), labels=data.columns.values)
    fig.colorbar(im, ax=ax)
    plt.title('Correlation matrix', fontsize=15)
    return im
    
    #Конец вашего кода

plot_corr_matrix(data)
````

### Шаг 3

Отфильтруйте параметры, у которых коэффициенты корреляции (абсолютные значения) по отношению к параметру ```"SalePrice"``` больше или равно ```0.4```. Затем передайте отфильтрованный ```pandas.DataFrame``` в функцию ```plot_corr_matrix()``` и получите рисунок, как показано в ожидаемых результатах.

````python
#Начало вашего кода
def plot_corr_matrix(data=None, figsize=None, threshold = 0.4):
    
    lst = []
    for i in data:
        if data[i].corr(data['SalePrice']) >= threshold:
            lst.append(i)
#     return data[lst]   
    corr_matrix = data[lst].corr()
#     return corr_matrix
    
    plt.rcParams["figure.figsize"] = figsize
    fig, ax = plt.subplots();
    im = ax.imshow(corr_matrix)
    
    plt.xticks(np.arange(len(data[lst].columns.values)), labels=data[lst].columns.values)
    plt.xticks(rotation=90)
    plt.yticks(np.arange(len(data[lst].columns.values)), labels=data[lst].columns.values)
    fig.colorbar(im, ax=ax)
    plt.title('Correlation matrix', fontsize=15)
    return im
#Конец вашего кода
plot_corr_matrix(data,figsize=(10,7))
````
