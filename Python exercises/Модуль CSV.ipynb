### 3. Создайте новые csv-файлы, содержащие данные только за тот же год. Пример: 2018.csv, 2017.csv и т. д.

````python
unique_years = set()
c = col_titles[0].index("Year")
for row in data:
    unique_years.add(row[c])
print(unique_years)
````

### Ожидаемый результат:
```
{'2013', '2018', '2017', '2014', '2016', '2015'}
```

````python

for year in unique_years:
    with open("{}.csv".format(year), "w") as ofile:
        csv_write = csv.writer(ofile)
        for row in data:
            if row[0] == year:
                csv_write.writerow(row)
    print("{}.csv is written".format(year))
````

### Ожидаемый результат:
